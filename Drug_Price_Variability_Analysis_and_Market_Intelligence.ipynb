# Section-2: Data Extraction Methodology


To investigate what drives pharmacy drug price differences across the U.S.‚Äîincluding the influence of geography and pharmacy behavior‚Äîwe developed an automated data extraction system using Google‚Äôs Gemini 2.0 Flash model. This pipeline was designed to accurately and efficiently convert unstructured visual data into structured datasets suitable for large-scale analysis.

Prescription pricing data was manually sourced in the form of approximately 150 screenshots from the publicly accessible GoodRx website. Each screenshot typically listed prices for a single drug across multiple pharmacies within a specific ZIP code. The images were organized into folders named after each ZIP code, allowing for location-aware processing. The extraction pipeline processed each folder in sequence, sending images through the Gemini API to transform visual content into structured text.

A carefully engineered prompt guided the model to return clean, machine-readable JSON containing the following key fields for each pharmacy listed in an image:
- **Drug name**
- **Location** (City followed by state abbreviation)
- **Pharmacy name** (as shown in the image)
- **Price** (numeric only, no currency symbols)
- **Prescription type** (e.g., dosage, tablet count)

To ensure API compliance, the system enforced automated rate limiting, capping requests to 15 per minute. If this limit was reached, the script automatically paused before resuming processing.

Each successful extraction was appended to a cumulative CSV file, allowing results to build across ZIP codes and drugs over multiple sessions. Any failures‚Äîsuch as malformed responses or parsing errors‚Äîwere logged with raw output for later inspection, ensuring the pipeline was robust and recoverable.

This approach enabled high-throughput, semi-automated data collection that accurately reflected real consumer-facing price data. The resulting dataset formed the foundation for analyzing both regional and pharmacy-specific pricing trends‚Äîultimately revealing that pharmacy choice, more than geography, plays a dominant role in price variability.

## 2.2 Define Utility Functions

````python
# Wait if rate limit is exceeded
def enforce_rate_limits():
    global REQUESTS_MADE, LAST_REQUEST_TIME
    if REQUESTS_MADE >= REQUESTS_PER_MINUTE:
        wait = max(0, 60 - (time.time() - LAST_REQUEST_TIME))
        if wait > 0:
            print(f"‚è≥ Waiting {wait:.1f}s to respect rate limit...")
            time.sleep(wait)
        REQUESTS_MADE = 0
        LAST_REQUEST_TIME = time.time()

# Error logger
def log_error(drug, message):
    with open(error_log_path, "a") as f:
        f.write(f"[{drug}] Error: {message}\n")
    print(f"‚ùå {drug}: {message}")

# Build structured prompt for image extraction
def build_prompt(drug_name):
    return f"""
Extract structured pricing data for the drug: {drug_name}

From each image, extract:
- Drug: {drug_name}
- Location (e.g., ‚ÄúNew York, NY‚Äù)
- Pharmacy
- Price (e.g., 995.00)
- Prescription

Return a clean JSON array only. No explanations or headers.
"""
````

## 2.3 Image Processing and Extraction Loop


````python
for folder in selected_folders:
    folder_path = os.path.join(base_path, folder)

    image_files = sorted([
        os.path.join(folder_path, f)
        for f in os.listdir(folder_path)
        if f.lower().endswith(".png")
    ])[:10]

    if not image_files:
        print(f"‚ö†Ô∏è Skipping {folder} (no images)")
        continue

    print(f"\nüì§ Processing folder: {folder}")
    folder_results = []
    prompt = build_prompt(folder)

    for idx, image_path in enumerate(image_files):
        enforce_rate_limits()
        print(f"üì∏ Image {idx + 1} sent...")

        try:
            image = Image.open(image_path)
            response = model.generate_content([prompt, image], stream=False)

            raw = response.candidates[0].content.parts[0].text
            cleaned = raw.strip().removeprefix("```json").removesuffix("```").strip()
            parsed = json.loads(cleaned)

            folder_results.extend(parsed)
            REQUESTS_MADE += 1
            LAST_REQUEST_TIME = time.time()

            print(f"‚úÖ Image {idx + 1} processed")

        except Exception as e:
            msg = f"Image {idx + 1} failed: {e}"
            log_error(folder, msg + f"\nüîé Raw output:\n{cleaned}\n")
            print(f"‚ùå {msg}")

    if folder_results:
        df_folder = pd.DataFrame(folder_results)
        df_all = pd.concat([df_all, df_folder], ignore_index=True)
        df_all.to_csv(csv_output_path, index=False)
        print(f"\n‚úÖ Saved {len(folder_results)} entries from {folder}")
    else:
        print(f"\n‚ùå No entries extracted from {folder}")

print(f"\nüéâ Extraction complete ‚Äî processed {len(selected_folders)} folders.")
````

While identifying the lowest price per drug is useful, it does not account for differences in packaging, dosage, or quantity. For a fair and consistent comparison across pharmacies, we now normalize prices based on **unit cost** and apply **Z-score standardization**. This helps identify which pharmacies offer systematically better pricing across a wide range of medications.


## 4.3 Normalized Drug Pricing Analysis

To ensure a fair, cross-drug comparison, we normalized prices by converting them to **unit costs** (e.g., per tablet, per pen, per vial), then computed a **Z-score** for each drug‚Äôs unit price across all pharmacies. This approach standardizes each price relative to its peers for that specific medication ‚Äî allowing us to compare affordability across drugs with vastly different base costs.

A lower average Z-score indicates that a pharmacy consistently offers **below-average prices**, regardless of how expensive the drugs are overall.

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Define unit counts for each drug (based on standard prescription dosage)
# This helps normalize price by quantity (e.g., per tablet, per pen)
unit_map = {
    "rivaroxaban": 60,
    "dapagliflozin": 30,
    "biktarvy": 30,
    "sitagliptin": 30,
    "apixaban": 60,
    "dulaglutide": 4,
    "insulin aspart": 1,
    "insulin glargine": 1,
    "semaglutide": 1,
    "empagliflozin": 30,
    "tirzepatide": 4,
    "risankizumab": 1,
    "dupilumab": 2,
    "etanercept": 4,
    "adalimumab": 2,
}

# 2. Add a column for units and calculate the unit price
# This removes dosage-based price inflation by converting to price per unit
df["Units"] = df["Drug"].map(unit_map)
df["Unit Price"] = df["Price"] / df["Units"]

# 3. Calculate Z-scores within each drug group
# This standardizes the price data so we can compare across drugs with vastly different price ranges
df["Z-Score"] = df.groupby("Drug")["Unit Price"].transform(
    lambda x: (x - x.mean()) / x.std()
)

# 4. Keep only pharmacies that carry 10 or more distinct drugs
# Ensures fair comparison: removes pharmacies with only 1‚Äì2 expensive or cheap drugs
pharmacy_counts = df.groupby("Pharmacy")["Drug"].nunique()
eligible_pharmacies = pharmacy_counts[pharmacy_counts >= 10].index
filtered_z_df = df[df["Pharmacy"].isin(eligible_pharmacies)]

# 5. Compute the average Z-score per pharmacy
# Lower Z-score = more affordable across all drugs carried
avg_zscore_per_pharmacy = (
    filtered_z_df.groupby("Pharmacy", as_index=False)["Z-Score"]
    .mean()
    .sort_values("Z-Score", ascending=True)  # Cheaper pharmacies appear first
)

# 6. Visualize the affordability using a barplot
# Affordability is based on relative (standardized) pricing across drugs
plt.figure(figsize=(12, 6))
sns.barplot(
    data=avg_zscore_per_pharmacy,
    x="Pharmacy",
    y="Z-Score",
    hue="Pharmacy",  # Hue used just for coloring
    dodge=False,
    palette="Spectral",
    legend=False
)
plt.axhline(0, linestyle='--', color='gray')  # Reference line at average affordability
plt.title("Pharmacy Affordability Score (Mean Z-Score of Unit Prices)", fontsize=14, weight="bold")
plt.ylabel("Average Z-Score (Lower is Better)")
plt.xlabel("Pharmacy")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# 7. Optional: View the final table with affordability scores
# Can be used in dashboards or summary tables
avg_zscore_per_pharmacy.reset_index(drop=True)
````

Key observations:

* **Hy-Vee** is the most affordable pharmacy in the dataset, with an average Z-score of **-1.20**, meaning it consistently offers unit prices well below the norm across all 15 drugs studied.
* **King Kullen**, **Nob Hill**, and **Raley‚Äôs** follow closely behind, each with Z-scores below **-0.9**, reflecting strong affordability across multiple drug classes.
* Large national chains like **CVS**, **Walgreens**, and **Walmart** fall near or slightly above the mean (Z-scores from **-0.2 to +0.5**), suggesting their pricing is **mixed** ‚Äî some drugs are competitively priced, others less so.
* **Costco**, somewhat surprisingly, ranks as the **least affordable** in this analysis with a Z-score of **+0.88**, indicating it consistently charges **higher-than-average unit prices** across the board.
* Pharmacies near the average line ‚Äî such as **Rite Aid**, **Publix**, and **Walgreens Specialty** ‚Äî offer **moderate affordability**, with prices generally close to the overall mean.


üìå **Why Do Some Smaller Chains Offer Lower Prices?**  
Pharmacies like Hy-Vee, King Kullen, Nob Hill, and Price Chopper ‚Äî all of which rank among the most affordable in this dataset ‚Äî are generally regional grocery-linked chains or independent operators. Their lower prices may stem from:

- Operating in less saturated, price-sensitive markets (e.g., Midwest, small cities)
- Grocery store integration that offsets pharmacy operating costs
- Flexible participation in coupon and discount networks
- Lower overhead compared to national chains
- Less entanglement with PBM(Pharmacy Benefit Managers) pricing structures, allowing more competitive cash pricing

This suggests that **smaller regional chains may provide hidden affordability advantages**, particularly for uninsured or self-pay patients.


üìå **Note on Costco‚Äôs High Z-Score**:  
While Costco is often known for offering competitive pricing on consumer goods and over-the-counter medications, its relatively high Z-score (+0.88) in this analysis may reflect the **nature of the drugs studied**. The selected medications are largely **specialty or chronic-use prescriptions** ‚Äî such as injectables for autoimmune conditions, advanced diabetes treatments, and HIV therapy ‚Äî rather than common generics or mass-market drugs like Tylenol or Lisinopril.

Costco‚Äôs pricing strategy may be less optimized for these specialized drugs, particularly if they‚Äôre not purchased in high volume or covered under standard pharmacy benefit contracts. This reinforces the idea that **affordability depends not only on the pharmacy chain but also on the specific drug class and use case.**

**Inference**: By standardizing for dosage and price range, this analysis reveals which pharmacies are **systematically more affordable** ‚Äî not just for one drug, but across a broad range of prescriptions. These findings are valuable for patients, providers, and policymakers seeking transparency and cost savings in pharmaceutical purchasing.


# Section 5: Hypothesis testing

## 5.2 Hypothesis Testing: Do Pharmacy Drug Price Differences Matter?

While our previous visualizations and Z-score rankings highlighted differences in affordability across pharmacies, we now turn to **formal statistical testing** to determine whether these pricing gaps are **statistically significant**.

We test the following hypothesis for each drug:

- **Null Hypothesis (H‚ÇÄ):** The mean drug prices across all pharmacies are equal.  
- **Alternative Hypothesis (H‚ÇÅ):** At least one pharmacy has a significantly different mean price.

To ensure fair and valid comparison, we include only those drugs that have **price entries from at least 9 different pharmacies**. This avoids noisy results due to sparse or unbalanced data.

By applying **one-way ANOVA** to each drug, we can rigorously assess whether the observed pricing variation between pharmacy chains is likely due to real market differences ‚Äî or just statistical noise.


````python
from scipy import stats

# Step 1: Filter drugs with at least 9 entries in each region
drug_region_counts = df.groupby(['Drug', 'Region']).size().unstack(fill_value=0)
valid_drugs = drug_region_counts[(drug_region_counts >= 9).all(axis=1)].index.tolist()
df_balanced = df[df['Drug'].isin(valid_drugs)]

# Step 2: Run ANOVA
anova_results = []

for drug in df_balanced['Drug'].unique():
    grouped_prices = [
        group['Price'].dropna()
        for name, group in df_balanced[df_balanced['Drug'] == drug].groupby('Region')
        if name is not None
    ]

    if len(grouped_prices) >= 2:
        f_stat, p_val = stats.f_oneway(*grouped_prices)
        anova_results.append({
            'Drug': drug,
            'F-statistic': f_stat,
            'p-value': p_val
        })

# Step 3: Create results DataFrame
df_anova = pd.DataFrame(anova_results)
df_anova['Significant (p < 0.05)'] = df_anova['p-value'] < 0.05
df_anova = df_anova.sort_values('p-value')

print(df_anova)
````

The results show that:

- **None of the drugs produced a statistically significant p-value (p < 0.05)**
- All p-values were well above 0.7 ‚Äî most exceeding 0.95
- This strongly supports the **null hypothesis**: there is **no meaningful difference in average drug price across regions**

This suggests that **regional location is not a major factor in price variation** for the drugs analyzed.


````python
# Set up the grid: 2 rows √ó 5 columns
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(22, 10))
axes = axes.flatten()  # Flatten to make iteration easier

# Plot each drug into a subplot
for i, drug in enumerate(valid_drugs):
    ax = axes[i]
    sns.boxplot(
        data=df_balanced[df_balanced['Drug'] == drug],
        x='Region',
        y='Price',
        palette='Set3',
        ax=ax,
        hue = 'Region',
        legend=False
    )
    ax.set_title(drug, fontsize=12, weight='bold')
    ax.set_xlabel("")
    ax.set_ylabel("Price (USD)")
    ax.tick_params(axis='x', rotation=45)

# Remove empty subplots if any
for j in range(len(valid_drugs), len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.suptitle("Price Distribution by Region for Top 10 Drugs", fontsize=16, weight='bold')
plt.tight_layout(rect=[0, 0, 1, 0.95])  # Leave space for main title
plt.show()
````

These boxplots visualize how drug prices vary across regions for each of the top 10 drugs in our dataset.

- For most drugs, the **price distributions across regions are highly similar**, with overlapping interquartile ranges and nearly identical medians.
- A few drugs (e.g., insulin glargine, biktarvy) show **greater spread** or more outliers, but this is not consistent enough to suggest meaningful regional effects.
- Overall, this visual confirms the results of our hypothesis testing: **regional location is not a major factor in price variation**


üìå **Note: Limitations - Scope and Generalizability**

While the ANOVA results and boxplot visualizations indicate no statistically significant price differences across regions for the selected drugs, a few important limitations should be noted:

- **Drug Selection Bias**:  
  The analysis is based on 15 high-impact, brand-name prescription drugs ‚Äî many of which are newer or used for chronic diseases (e.g., diabetes, autoimmune conditions). These findings may not hold for lower-cost generics or over-the-counter medications.

- **ZIP Code Sampling**:  
  Although ZIP codes were selected to represent different regions, urban/rural settings, and cost-of-living tiers, the sample is still limited (~10 ZIPs total). A larger, more granular set of ZIP codes may reveal finer geographic trends.

- **Pharmacy Representation**:  
  Not all pharmacy chains were equally represented in every region. For example, some affordable regional pharmacies may only exist in one part of the country. This can blur true regional effects if price differences are actually pharmacy-driven.

- **Data Source Constraints**:  
  Prices were extracted from GoodRx, which reflects **cash prices**, not insurance-adjusted costs. Patients using insurance or manufacturer coupons may experience very different pricing dynamics.

- **Time Sensitivity**:  
  Prices were captured at a single point in time. Drug pricing is dynamic and may fluctuate due to supply chain issues, manufacturer policy changes, or regulatory shifts.

These limitations suggest that while **pharmacy-specific pricing behavior** is likely a more powerful driver than geography ‚Äî especially for this drug group ‚Äî the conclusions should be interpreted with appropriate context and caution.


## 5.2 Hypothesis Testing: Do Pharmacy Drug Price Differences Matter?

So far, we've seen that some pharmacies consistently offer lower prices across drugs. But are these price differences **statistically significant**, or could they be due to random sampling noise?

To answer this, we apply a one-way **ANOVA (Analysis of Variance)** test across pharmacies for each drug.

**Hypotheses Tested:**

- **Null Hypothesis (H‚ÇÄ):** The mean price of a given drug is equal across all pharmacies.
- **Alternative Hypothesis (H‚ÇÅ):** At least one pharmacy has a significantly different mean price.

**Data Filtering:**

To ensure the analysis is statistically reliable, we included only drugs that:
- Appear in **at least 3 different pharmacies**, and
- Have **5 or more data points per pharmacy**.

This avoids skewed results from underrepresented groups.

````python
# Step 1: Filter drugs that appear in ‚â•3 pharmacies with ‚â•5 entries each
drug_pharmacy_counts = df.groupby(['Drug', 'Pharmacy']).size().unstack(fill_value=0)
valid_drugs = drug_pharmacy_counts[(drug_pharmacy_counts >= 5).sum(axis=1) >= 3].index.tolist()
df_pharmacy_filtered = df[df['Drug'].isin(valid_drugs)]

# Step 2: Run one-way ANOVA to compare prices across pharmacies for each drug
anova_pharmacy_results = []

for drug in df_pharmacy_filtered['Drug'].unique():
    grouped_prices = [
        group['Price'].dropna()
        for _, group in df_pharmacy_filtered[df_pharmacy_filtered['Drug'] == drug].groupby('Pharmacy')
        if len(group) >= 5
    ]

    if len(grouped_prices) >= 2:
        try:
            f_stat, p_val = stats.f_oneway(*grouped_prices)
        except Exception:
            f_stat, p_val = float('nan'), float('nan')

        anova_pharmacy_results.append({
            'Drug': drug,
            'F-statistic': f_stat,
            'p-value': p_val
        })

# Step 3: Format the result table
df_pharmacy_anova = pd.DataFrame(anova_pharmacy_results)
df_pharmacy_anova['Significant (p < 0.05)'] = df_pharmacy_anova['p-value'] < 0.05
df_pharmacy_anova = df_pharmacy_anova.sort_values('p-value')

# Clean formatting for display
df_pharmacy_anova['F-statistic'] = df_pharmacy_anova['F-statistic'].round(2)
df_pharmacy_anova['p-value'] = df_pharmacy_anova['p-value'].apply(lambda x: f"{x:.2e}" if pd.notnull(x) else "NaN")
df_pharmacy_anova.reset_index(drop=True, inplace=True)

# Show result
df_pharmacy_anova
````

The results show that:

- **All but one drug showed statistically significant differences** (p < 0.05) in mean price across pharmacies.
- Several drugs had **very high F-statistics**, even returning \`inf`, suggesting near-zero within-group variance and large between-group gaps.
- **Adalimumab** was excluded from statistical conclusions due to insufficient or uniform data.

These results clearly reject the null hypothesis for nearly all drugs ‚Äî confirming that **pharmacy choice strongly impacts drug pricing**.
